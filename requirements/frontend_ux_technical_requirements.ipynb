### 4.5 AgeBadge

Color-coded age display based on branch-specific thresholds:

| Condition | Color | Tailwind Class |
|-----------|-------|----------------|
| Not stale (within threshold) | Green/Default | `text-green-600 bg-green-50` |
| Approaching threshold (75%+) | Amber/Warning | `text-amber-600 bg-amber-50` |
| Past threshold (stale) | Red/Critical | `text-red-600 bg-red-50` |

**Logic:**
```typescript
// For Main PRs: threshold = 3 days
// For Feature PRs: threshold = 2 days
function getAgeColor(ageDays: number, thresholdDays: number): string {
  if (ageDays >= thresholdDays) return "critical";      // Red
  if (ageDays >= thresholdDays * 0.75) return "warning"; // Amber
  return "normal";                                       // Green
}
```

## 5. Data Flow & State Management

### 5.1 TanStack Query (React Query) Setup

```typescript
// hooks/usePullRequests.ts
export function usePullRequests(filters: PRFilters) {
  return useQuery({
    queryKey: ['pull-requests', filters],
    queryFn: () => fetchPullRequests(filters),
    refetchInterval: 2 * 60 * 1000,  // Auto-refetch every 2 minutes
    staleTime: 60 * 1000,             // Data considered fresh for 1 min
  });
}

// hooks/usePRDetail.ts
export function usePRDetail(prNumber: number | null) {
  return useQuery({
    queryKey: ['pull-request-detail', prNumber],
    queryFn: () => fetchPRDetail(prNumber!),
    enabled: !!prNumber,  // Only fetch when drawer is open
  });
}
```

### 5.2 State Flow

```
User Action          →  State Change              →  UI Update
─────────────────────────────────────────────────────────────────
Page load            →  fetch PRs (all)           →  Render table
Click tab            →  Update filter state       →  Filter table rows
Click column header  →  Update sort state         →  Re-sort table rows
Click row            →  Set selectedPR            →  Open drawer + fetch detail
Click refresh        →  Invalidate query cache    →  Refetch + update table
Auto-refresh timer   →  Background refetch        →  Update table seamlessly
Close drawer         →  Set selectedPR = null     →  Close panel
```

### 5.3 Local State (Component-level)

| State | Type | Default | Persisted? |
|-------|------|---------|------------|
| `activeTab` | `"all" \| "main" \| "feature"` | `"all"` | Session (URL param) |
| `sortBy` | `"age" \| "author" \| "reviewers"` | `"age"` | No |
| `sortOrder` | `"asc" \| "desc"` | `"desc"` | No |
| `selectedPR` | `number \| null` | `null` | No |

## 6. TypeScript Types

```typescript
// types/pull-request.ts

interface PullRequest {
  number: number;
  title: string;
  author: {
    username: string;
    avatar_url: string;
  };
  base_branch: string;
  head_branch: string;
  branch_type: "main" | "feature";
  created_at: string;
  age: {
    days: number;
    hours: number;
    display: string;
    is_stale: boolean;
    threshold_days: number;
  };
  active_reviewers_count: number;
  code_owner_status: {
    required: boolean;
    approved: boolean;
  } | null;
  html_url: string;
  labels: string[];
}

interface PullRequestDetail extends PullRequest {
  body: string;
  active_reviewers: {
    username: string;
    avatar_url: string;
    state: "APPROVED" | "CHANGES_REQUESTED" | "COMMENTED";
  }[];
  code_owner_status: {
    required: boolean;
    approved: boolean;
    owners: {
      username: string;
      has_approved: boolean;
    }[];
  } | null;
  files_changed: {
    total: number;
    additions: number;
    deletions: number;
  };
}

interface PRListResponse {
  total_count: number;
  filters_applied: {
    branch_type: string;
    sort_by: string;
    sort_order: string;
  };
  pull_requests: PullRequest[];
}

interface PRFilters {
  branch_type: "all" | "main" | "feature";
  sort_by: "age" | "author" | "reviewers";
  sort_order: "asc" | "desc";
}
```

## 8. Interaction Patterns

### 8.1 Table Row Click → Drawer (Overlay Pattern)

```
1. User clicks a PR row
2. Selected row gets highlighted background
3. Semi-transparent backdrop appears over the entire page
4. Drawer slides in from right (300ms ease-out animation)
5. Body scroll is LOCKED (page scroll position stays unchanged)
6. Table remains visible underneath but is dimmed and NOT interactive
7. Drawer shows skeleton while detail API loads
8. Detail data populates
9. Drawer content is independently scrollable (overflow-y: auto)
10. Focus is trapped inside the drawer (Tab cycles within drawer only)
```

**Closing the drawer:**
```
1. User clicks backdrop / presses Escape / clicks X button
2. Drawer slides out to the right (200ms ease-in)
3. Backdrop fades out
4. Body scroll is UNLOCKED (restored to original position)
5. Focus returns to the table row that was clicked
6. Table becomes interactive again
```

### 8.2 Refresh Behavior

```
Manual Refresh:
1. User clicks refresh button
2. Button shows spinner
3. Data refetches in background
4. Table updates smoothly (no full re-render)
5. "Last updated" timestamp resets

Auto Refresh (every 2 min):
1. Background refetch (no visual indicator)
2. If data changed, table updates seamlessly
3. If drawer is open, drawer data also refreshes
```

### 8.3 Sort Interaction

```
1. User clicks column header (e.g., "Age")
2. First click: sort descending (oldest first)
3. Second click: sort ascending
4. Arrow icon shows current sort direction
5. Only one column can be sorted at a time
```

### 8.4 Loading States

| State | Display |
|-------|---------|
| Initial load | Full skeleton table (5 skeleton rows) |
| Refetching | Subtle loading bar at top (no skeleton) |
| Drawer loading | Skeleton blocks in drawer |
| Error | Error message with retry button |

## 10. API Integration

### 10.1 API Client Configuration

```typescript
// api/client.ts
const API_BASE_URL = import.meta.env.VITE_API_URL || 'http://localhost:8000';

export async function apiClient<T>(endpoint: string): Promise<T> {
  const response = await fetch(`${API_BASE_URL}${endpoint}`);
  if (!response.ok) {
    throw new Error(`API error: ${response.status}`);
  }
  return response.json();
}
```

### Dockerfile (Frontend — Nginx serving static files)

```dockerfile
FROM node:20-alpine AS builder
WORKDIR /app
COPY package*.json ./
RUN npm ci
COPY . .
RUN npm run build

FROM nginx:alpine
COPY --from=builder /app/dist /usr/share/nginx/html
COPY nginx.conf /etc/nginx/conf.d/default.conf
EXPOSE 80
```

### nginx.conf

```nginx
server {
    listen 80;
    root /usr/share/nginx/html;
    index index.html;

    # SPA fallback
    location / {
        try_files $uri $uri/ /index.html;
    }

    # Proxy API calls to backend
    location /api/ {
        proxy_pass http://backend:8000;
        proxy_set_header Host $host;
        proxy_set_header X-Real-IP $remote_addr;
    }
}
```

> **Note:** In production (Linux deployment), the frontend Nginx container proxies `/api/` to the backend container. Both run via `docker-compose` (see backend doc for compose file).
