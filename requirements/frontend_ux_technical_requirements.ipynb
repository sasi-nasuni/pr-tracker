### 4.5 AgeBadge

Color-coded age display based on branch-specific thresholds:

| Condition | Color | Tailwind Class |
|-----------|-------|----------------|
| Not stale (within threshold) | Green/Default | `text-green-600 bg-green-50` |
| Approaching threshold (75%+) | Amber/Warning | `text-amber-600 bg-amber-50` |
| Past threshold (stale) | Red/Critical | `text-red-600 bg-red-50` |

**Logic:**
```typescript
// For Main PRs: threshold = 3 days
// For Feature PRs: threshold = 2 days
function getAgeColor(ageDays: number, thresholdDays: number): string {
  if (ageDays >= thresholdDays) return "critical";      // Red
  if (ageDays >= thresholdDays * 0.75) return "warning"; // Amber
  return "normal";                                       // Green
}
```

## 5. Data Flow & State Management

### 5.1 TanStack Query (React Query) Setup

```typescript
// hooks/usePullRequests.ts — Phase 2 (client-side filtering)
export function usePullRequests() {
  return useQuery({
    queryKey: ['pull-requests'],              // Single stable key — no filter dependency
    queryFn: () => fetchPullRequests(),       // No filter params sent to backend
    refetchInterval: 2 * 60 * 1000,          // Auto-refetch every 2 minutes
    staleTime: 60 * 1000,                     // Data considered fresh for 1 min
  });
}

// Client-side filtering & sorting via useMemo
export function useFilteredPRs(data: PullRequest[], filters: PRFilters) {
  return useMemo(() => {
    let result = data;

    // Filter by branch type
    if (filters.branch_type !== 'all') {
      result = result.filter(pr => pr.branch_type === filters.branch_type);
    }

    // Sort
    result = [...result].sort((a, b) => {
      // Sort logic based on filters.sort_by and filters.sort_order
    });

    return result;
  }, [data, filters]);
}

// hooks/usePRDetail.ts (unchanged)
export function usePRDetail(prNumber: number | null) {
  return useQuery({
    queryKey: ['pull-request-detail', prNumber],
    queryFn: () => fetchPRDetail(prNumber!),
    enabled: !!prNumber,  // Only fetch when drawer is open
  });
}
```

> **Phase 2 Change:** The `queryKey` no longer includes filter state. Filters are applied client-side via `useMemo`, making tab switches and sort changes **instant** with no network requests. The backend still supports filter params — the frontend simply doesn't use them anymore.

### 5.2 State Flow (Phase 2)

```
User Action          →  State Change              →  UI Update
─────────────────────────────────────────────────────────────────
Page load            →  fetch all PRs             →  Render table (filtered by default tab)
Click tab            →  Update filter state       →  useMemo recalculates → instant re-render
Click column header  →  Update sort state         →  useMemo recalculates → instant re-render
Click row            →  Set selectedPR            →  Open drawer + fetch detail
Click refresh        →  Invalidate query cache    →  Refetch + update table
Auto-refresh timer   →  Background refetch        →  Update table seamlessly
Close drawer         →  Set selectedPR = null     →  Close panel
```

### 5.3 Local State (Component-level)

| State | Type | Default | Persisted? |
|-------|------|---------|------------|
| `activeTab` | `"all" \| "main" \| "feature"` | `"all"` | Session (URL param) |
| `sortBy` | `"age" \| "author" \| "reviewers"` | `"age"` | No |
| `sortOrder` | `"asc" \| "desc"` | `"desc"` | No |

## 6. TypeScript Types

```typescript
// types/pull-request.ts

interface TeamApprovals {
  count: number;
  required: number;
  sufficient: boolean;
  reviewers: string[];  // Names shown in tooltip on hover
}

interface UnresolvedThread {
  commenter: string;        // GitHub username
  file_path: string | null; // File the comment is on (null for PR-level)
  comment_snippet: string;  // First ~100 chars of the comment
  url: string;              // Direct link to comment on GitHub
}

interface PullRequest {
  number: number;
  title: string;
  author: {
    username: string;
    avatar_url: string;
  };
  base_branch: string;
  head_branch: string;
  branch_type: "main" | "feature";
  created_at: string;
  age: {
    days: number;
    hours: number;
    display: string;
    is_stale: boolean;
    threshold_days: number;
  };
  active_reviewers_count: number;
  team_approvals: TeamApprovals;              // Phase 2
  unresolved_comment_count: number;           // Phase 2
  code_owner_status: {
    required: boolean;
    approved: boolean;
  } | null;
  html_url: string;
  labels: string[];
}

interface PullRequestDetail extends PullRequest {
  body: string;
  active_reviewers: {
    username: string;
    avatar_url: string;
    state: "APPROVED" | "CHANGES_REQUESTED" | "COMMENTED";
  }[];
  code_owner_status: {
    required: boolean;
    approved: boolean;
    owners: {
      username: string;
      has_approved: boolean;
    }[];
  } | null;
  files_changed: {
    total: number;
    additions: number;
    deletions: number;
  };
  // Phase 2 detail additions:
  team_approvals: TeamApprovals;              // Full breakdown (same shape, used for detail section)
  unresolved_threads: UnresolvedThread[];     // Full thread details for detail panel
}

interface PRListResponse {
  total_count: number;
  filters_applied: {                          // Still returned by backend
    branch_type: string;
    sort_by: string;
    sort_order: string;
  };
  pull_requests: PullRequest[];
}

interface PRFilters {
  branch_type: "all" | "main" | "feature";
  sort_by: "age" | "author" | "reviewers";
  sort_order: "asc" | "desc";
}
```

## 8. Interaction Patterns

### 8.1 Table Row Click → Drawer (Overlay Pattern)

```
1. User clicks a PR row
2. Selected row gets highlighted background
3. Semi-transparent backdrop appears over the entire page
4. Drawer slides in from right (300ms ease-out animation)
5. Body scroll is LOCKED (page scroll position stays unchanged)
6. Table remains visible underneath but is dimmed and NOT interactive
7. Drawer shows skeleton while detail API loads
8. Detail data populates
9. Drawer content is independently scrollable (overflow-y: auto)
10. Focus is trapped inside the drawer (Tab cycles within drawer only)
```

**Closing the drawer:**
```
1. User clicks backdrop / presses Escape / clicks X button
2. Drawer slides out to the right (200ms ease-in)
3. Backdrop fades out
4. Body scroll is UNLOCKED (restored to original position)
5. Focus returns to the table row that was clicked
6. Table becomes interactive again
```

### 8.2 Refresh Behavior

```
Manual Refresh:
1. User clicks refresh button
2. Button shows spinner
3. Data refetches in background
4. Table updates smoothly (no full re-render)
5. "Last updated" timestamp resets

Auto Refresh (every 2 min):
1. Background refetch (no visual indicator)
2. If data changed, table updates seamlessly
3. If drawer is open, drawer data also refreshes
```

### 8.3 Sort Interaction

```
1. User clicks column header (e.g., "Age")
2. First click: sort descending (oldest first)
3. Second click: sort ascending
4. Arrow icon shows current sort direction
5. Only one column can be sorted at a time
```

### 8.4 Loading States

| State | Display |
|-------|---------|
| Initial load | Full skeleton table (5 skeleton rows) |
| Refetching | Subtle loading bar at top (no skeleton) |
| Drawer loading | Skeleton blocks in drawer |
| Error | Error message with retry button |

## 10. API Integration

### 10.1 API Client Configuration

```typescript
// api/client.ts
const API_BASE_URL = import.meta.env.VITE_API_URL || 'http://localhost:8000';

export async function apiClient<T>(endpoint: string): Promise<T> {
  const response = await fetch(`${API_BASE_URL}${endpoint}`);
  if (!response.ok) {
    throw new Error(`API error: ${response.status}`);
  }
  return response.json();
}
```

### 17.5 Updated Client-Side Filtering Logic

```typescript
// App.tsx — updated useMemo with repo filter
const filteredPRs = useMemo(() => {
  let result = allPRs;

  // Filter by repository
  if (filters.repository !== "all") {
    result = result.filter((pr) => pr.repository === filters.repository);
  }

  // Filter by branch type
  if (filters.branch_type !== "all") {
    result = result.filter((pr) => pr.branch_type === filters.branch_type);
  }

  // Sort
  const reverse = filters.sort_order === "desc";
  result = [...result].sort((a, b) => {
    let cmp = 0;
    if (filters.sort_by === "age") {
      cmp = (a.age.days * 24 + a.age.hours) - (b.age.days * 24 + b.age.hours);
    } else if (filters.sort_by === "author") {
      cmp = a.author.username.localeCompare(b.author.username);
    } else if (filters.sort_by === "reviewers") {
      cmp = a.active_reviewers_count - b.active_reviewers_count;
    }
    return reverse ? -cmp : cmp;
  });

  return result;
}, [allPRs, filters]);

// Compute repo list from full dataset
const repos = useMemo(() => {
  const repoMap = new Map<string, number>();
  for (const pr of allPRs) {
    repoMap.set(pr.repository, (repoMap.get(pr.repository) || 0) + 1);
  }
  return Array.from(repoMap, ([name, count]) => ({ name, count }))
    .sort((a, b) => b.count - a.count);
}, [allPRs]);

// Tab counts reflect both filters
const counts = useMemo(() => {
  const repoFiltered = filters.repository === "all"
    ? allPRs
    : allPRs.filter((pr) => pr.repository === filters.repository);
  return {
    all: repoFiltered.length,
    main: repoFiltered.filter((pr) => pr.branch_type === "main").length,
    feature: repoFiltered.filter((pr) => pr.branch_type === "feature").length,
  };
}, [allPRs, filters.repository]);
```
